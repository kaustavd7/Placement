```python
import csv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('training.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Trim'].fillna(method = 'bfill', inplace = True)
df['Trim'].isnull().sum()
```

```python
df['SubModel'].fillna(df['SubModel'].mode()[0], inplace=True)
df['Color'].fillna(df['Color'].mode()[0], inplace=True)
df['Transmission'].fillna(df['Transmission'].mode()[0], inplace=True)
```

```python
sns.boxplot(df['WheelTypeID'])
```

```python
df['WheelTypeID'].isnull().sum()
```

```python

df['WheelTypeID'].fillna(df['WheelTypeID'].mean(), inplace = True)
```

```python
sns.boxplot(df['WheelType'])
```

```python
df['WheelType'].fillna(df['WheelType'].mode()[0], inplace = True)
```

```python
df['Nationality'].fillna(df['Nationality'].mode()[0], inplace = True)
df['Size'].fillna(method = 'bfill', inplace = True)
df['TopThreeAmericanName'].fillna(method = 'bfill', inplace = True)
```

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
for col in df.columns:
    if col.startswith('MMR'):
        plt.figure(figsize=(8, 5))
        plt.hist(df[col].dropna(), bins=30, edgecolor='black', alpha=0.7)  # Dropping NaN values for plotting
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()
```

```python
for col in df.columns:
    if col.startswith('MMR'):
        # Fill missing values with the median of the column
        df[col].fillna(df[col].median(), inplace=True)
```

```python
from sklearn.feature_selection import chi2
from sklearn.preprocessing import LabelEncoder
import numpy as np

# Encode the 'Trim' column (categorical feature)
le = LabelEncoder()
X_encoded = le.fit_transform(df['Trim']).reshape(-1, 1)  # Reshape to (n, 1)

# Encode the target variable 'IsBadGuy'
y_encoded = le.fit_transform(df['IsBadBuy'])  # Already 1D, no need for reshape

# Compute chi-squared statistic
chi2_stat, p_val = chi2(X_encoded, y_encoded)
print(f"Chi2 Stat: {chi2_stat[0]}, P-value: {p_val[0]}")
```

```python
from sklearn.feature_selection import mutual_info_classif

# Example: Single feature and prediction column
# feature = df['SubModel'].values.reshape(-1, 1)
feature = df['Trim']
target = df['IsBadBuy']

# Encode categorical data
feature_encoded = LabelEncoder().fit_transform(feature)
target_encoded = LabelEncoder().fit_transform(target)

# # Compute mutual information
# mi = mutual_info_classif(feature, target, discrete_features='auto')
# print(f"Mutual Information: {mi[0]}")
mi = mutual_info_classif(feature_encoded.reshape(-1, 1), target_encoded, discrete_features=True)
print(f"Mutual Information: {mi[0]}")
```
